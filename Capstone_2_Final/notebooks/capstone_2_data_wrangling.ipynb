We hypothesize that grades should be one of the most important features in determining whether a students has paid tutoring.

A grade of zero appears to be a substitute for null values and infeasible (considering no students in the first period (G1) had a grade of zero), and so we should remove any rows containing a value of zero. 

```python
#Removing any rows with values of zero in the G2 feature
data.drop(data.loc[data['G2']==0].index, inplace=True)
data.G2.value_counts()
```

```python
data.G3.value_counts()
```

Similarly to G2, we should remove any rows containing a value of zero in G3 (third period) as well. 

```python
#Removing any rows with values of zero in the G3 feature
data.drop(data.loc[data['G3']==0].index, inplace=True)
data.G3.value_counts()
```
