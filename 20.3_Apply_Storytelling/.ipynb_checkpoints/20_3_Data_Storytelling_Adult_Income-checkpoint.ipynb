```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('adult.csv')
data.head()
```

```python
data = pd.read_csv('adult.csv')

data = data[data.age >= 22]
data = data[data.age <= 28]

rows = ['HS-grad', 'Some-college', 'Bachelors', 'Masters']
data = data[data.education.isin(rows)]
data.education.value_counts()
```

```python
data = data.loc[:, ['education', 'income']]
data
```

```python
df = pd.get_dummies(data, columns=data.columns[1:])
df['income_>50K'].replace(to_replace=0, value=1, inplace=True)
df2 = df.groupby(by='education').sum()

index = ['HS-grad', 'Some-college', 'Bachelors', 'Masters']
df2 = df2.reindex(index)
df2
```

```python
df2.plot.bar()
plt.ylabel('Count of Adults')
plt.title('Income Group of Adults in the Age Group of 22 to 28')
```
