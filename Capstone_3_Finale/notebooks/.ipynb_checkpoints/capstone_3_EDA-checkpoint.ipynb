```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import pylab as py
```

```python
df = pd.read_csv('../data/city_temperature_cleaned.csv')
```

```python
df.head()
```

In order to measure and quantify the increase in average temperature throughout the years, we will use the temperature anomaly method where we calculate the change in temperature from a particular reference temperature.

In our case, we will use the average monthly temperature in 1995 as the reference point for calculating the change in monthly average temperature for future years.

```python
#Calculating the 1995 monthly temperatures to use as the reference temperatures
df_1995 = df[df.Year == 1995]
monthly_avg_1995 = df_1995.groupby(df_1995.Month).mean()
monthly_avg_1995
```

We then want to group our dataset by month and year through aggregating by the mean of AvgTemperature in order to obtain the average monthly temperatures for 1996-2019 in order to compare with the 1995 reference temperatures.

```python
#Create a temporary copy of the original dataset
temp = df
temp.head()
```

```python
#Group by year & month, and then aggregate by mean to obtain monthly average temperature

#Set day as 1 for easier grouping by month through index
temp.Day = 1

date = pd.to_datetime(temp[['Year', 'Month', 'Day']])
temp = pd.concat([temp, date], axis=1)
temp.rename(columns={0:'Date_By_Month'}, inplace=True)
temp.set_index('Date_By_Month', inplace=True)
temp.index = pd.to_datetime(temp.index)

global_monthly_avg = temp.groupby(temp.index).mean()
global_monthly_avg
```

We can now subtract each month's average temperature by the reference temperature (1995 monthly average temperatures) to get the temperature anomaly values.

```python
#Calculate temperature anomaly
monthly_avg_1995 = monthly_avg_1995.append([monthly_avg_1995]*24, ignore_index=True)
global_monthly_avg['RefTemperature'] = np.array(monthly_avg_1995.AvgTemperature)
global_monthly_avg['Temp_Anomaly'] = global_monthly_avg.AvgTemperature - global_monthly_avg.RefTemperature
global_monthly_avg
```

We ideally want the Temp_Anomaly distribution to normally distributed, and so we can perform checks for hints of normality.

```python
#Check through histogram
sns.histplot(global_monthly_avg.Temp_Anomaly, bins=50)
plt.title('Monthly Temperature Anomaly Distribution')
```

This appears to resemble a normal distribution. We can then check for the normality through a qq-plot.

```python
sm.qqplot(global_monthly_avg.Temp_Anomaly, line ='45')
plt.title('Monthly Temperature Anomaly QQ-Plot')
py.show()
```

The distribution of Temp_Anomaly is very close to the 45 degree line, and so we can safely conclude that the data is normally distributed.

```python
#Reset index so that the date column is carried over to the csv file
df_month = global_monthly_avg
df_month.reset_index(inplace=True)
df_month
```
